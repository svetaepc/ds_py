```python
from sympy import *

x = symbols('x')
f =  x ** 3 - 50*x
g = -x**4 + 88*x**2 - 241

h = f - g

plot(f, g, (x, -15, 15))
```

```python
k = solve(h)
[[el.n().round(2),f.subs(x,el).n().round(2)] for el in k]
```
